```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl as ol
plt.rcParams['font.family'] = 'SimHei'
```

```python

def shanchuduoyu(df,path2):
    df["Building type"]=df["Building type"].astype(str).str.strip()
    df["A"]=df["A"].astype(str).str.strip()
    df["B"]=df["B"].astype(str).str.strip()
    df.to_excel(path2,index=False)
##将建造类型改为独热编码
def onehotsome(df,path2,path3):
    aqs=ol.load_workbook(path2)
    a=aqs.active
    # print(a.cell(2,5).value)
    # print(a.cell(2,6).value)
    # print("mid-rise")
    #print(a.cell(2,7).value)
    for i in range(1,df.shape[0]):
        for j in range(5,8):
            if a.cell(i,j).value=="multi-story":
                a.cell(i,8,1)
            elif a.cell(i,j).value=="mid-rise":
                a.cell(i,9,1)
            elif a.cell(i,j).value=="high-rise":
                a.cell(i,10,1)
            elif a.cell(i,j).value=="super high-rise":
                a.cell(i,11,1)
    aqs.save(path3)

```

```python
book=pd.read_excel(path3)
# print(book.columns)

def queshizhi(book):
    book["multi-story"]=book["multi-story"].fillna(0)
    book["mid-rise"]=book["mid-rise"].fillna(0)
    book["high-rise"]=book["high-rise"].fillna(0)
    book["super high-rise"]=book["super high-rise"].fillna(0)
 

```

```python
def onehottype(app):
    appdumpro= pd.get_dummies(app["property type"], dummy_na=True)
    app=pd.concat([app,appdumpro],axis=1)
    return app
def shanchulie(app):
    app=app.drop(columns=[np.nan,"property type"])
    return app
def process(path1,path2,path3):
    df=pd.read_excel(path1) 
    shanchuduoyu(df,path2)
    onehotsome(df,path2,path3)
    book=pd.read_excel(path3)
    queshizhi(book)
    app=book.drop(columns=["Building type","A","B"])
    app=queshi(app)
    app=xiugaitype(app)
    app=onehottype(app)
    app=shanchulie(app)
    return app
def codegroup(df,str1,path):
    dfs={}
    for i in df[str1].unique():
        dfs[i]=df[df[str1]==i]
    for i,j in dfs.items():
        j=j.drop(columns=["adcode"])
        j.to_excel(path+"//{}.xlsx".format(i),index=False)

        
path1="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//Appendix 1.xlsx"
path2="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//output.xlsx"
path3="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//Appendixchuli1.xlsx"
path="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//城市1不同地区"
process(path1,path2,path3)
codegroup(process(path1,path2,path3),"adcode",path)

path4="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//Appendix 2.xlsx"
path5="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//output2.xlsx"
path6="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//Appendixchuli2.xlsx"
path0="C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//城市2不同地区"

codegroup(process(path4,path5,path6),"adcode",path0)

# appdumadcode= pd.get_dummies(app["adcode"], dummy_na=True)
# app=pd.concat([app,appdumadcode],axis=1)

# app=app.drop(columns=['property type','adcode'])

# app=app.drop(columns=[np.nan])
# print(app.columns)
# print(app.dtypes)
# print(app.shape)

# app.to_excel('C://Users//Dell//Desktop//2024“ShuWei Cup”_Problem//2024_“ShuWei Cup”D_Problem//later.xlsx', index=False)
```
